```python
# Install Yahoo finance python package
# pip install yfinance
# Library imports
import yfinance as yf
import pandas as pd
sp = yf.download("^GSPC", start="2025-07-30", end = "2026-03-11")
# Data frame comes in Wide form, drop ticker header (all SP500)
sp.columns = sp.columns.droplevel(1)
#fix index
sp = sp.reset_index()
#fix names
sp.columns.name = None
#set index to date time for joining on date
sp['Date'] = pd.to_datetime(sp['Date'])
sp = sp.set_index('Date')
sp.head()
```

```python
# Create day change percentage column to find volatile days for volatity clustering.
sp['Day Change %'] = (sp['Close'] - sp['Open']) / sp['Open'] * 100
# Create overnight change for target var
sp['Overnight Change %'] = (sp['Open'] - sp['Close'].shift(1)) / sp['Close'].shift(1) * 100
# drop first observation since it has no overnight change
sp = sp.dropna(subset=['Overnight Change %'])
sp.head()
```

```python
#Save to csv for use in notebook
#sp.to_csv('SP500_data.csv')
#Save to parquet for use in notebook and maintain data types
sp.to_parquet('../data/SP500_data.parquet')
```
